#### Pegando os campos

```python
v_mat =         pagina_principal["data"]["content"][0]["matricula"]
v_nome =        pagina_principal["data"]["content"][0]["nome"]
v_cpf_part =    pagina_principal["data"]["content"][0]["cpf"]
v_mes =         pagina_principal["data"]["content"][0]["mes"]
v_ano =         pagina_principal["data"]["content"][0]["ano"]
v_orgao =       pagina_principal["data"]["content"][0]["orgao"]
v_bruto =       pagina_principal["data"]["content"][0]["remuneracaoBruta"]
v_liquido =     pagina_principal["data"]["content"][0]["remuneracaoLiquida"]

type(v_mat)
```

```python
colunas = ['matricula', 'nome', 'cpf_part', 'mes', 'ano','cargo', 'orgao', 'valor_bruto', 'valor_liquido']
df = pd.DataFrame(columns=colunas)
```

```python
for num_pag in range(3):

    params_0 = {'mes': 3, 'ano':2023, 'page': num_pag}
    req = re.get(url, params=params_0)

    pag = json.loads(req.text)

    dados_pagina = []

    #interando nos elementos das paginas
    

    try:
        for num_elem in range(20):
            elemento = pag["data"]["content"][num_elem]
            data = {
                'matricula': elemento["matricula"],
                'nome': elemento["nome"],
                'cpf_part': elemento["cpf"],
                'mes': elemento["mes"],
                'ano': elemento["ano"],
                'cargo': elemento["cargo"],
                'orgao': elemento["orgao"],
                'valor_bruto': elemento["remuneracaoBruta"],
                'valor_liquido': elemento["remuneracaoLiquida"]
            }
            dados_pagina.append(data)
            
            df_temp = pd.DataFrame(dados_pagina)
            df = pd.concat([df, df_temp])
    except IndexError:
        break  # Se não houver mais elementos na página, saia do loop

df.to_csv(f'Bases/servidores_RR_teste.csv', index=False)
```
